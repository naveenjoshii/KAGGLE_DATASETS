```python
df1 = data.drop('chas',axis = 1)
df1.drop('ID',axis = 1, inplace =True)
```

```python
X = data[['crim', 'zn', 'indus', 'nox', 'rm', 'age', 'dis', 'rad','tax', 'ptratio', 'black','lstat']]
y = data['medv']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 8)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
predictions = lm.predict(X_test)
plt.scatter(y_test,predictions, alpha=0.8)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
